## Analysing Data of 20 students based on there performance .

#### importing Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

#### Creating Data of 20 students 

```python
np.random.seed(42) # for same random number each time .
```

```python
students_name=[f"Student{i+1}" for i in range(20)]
math_grades = np.random.randint(60, 101, 20)
science_grades = np.random.randint(60, 101, 20)
english_grades = np.random.randint(60, 101, 20)
```

#### Implementing the above data using pandas

```python
df =pd.DataFrame({
    'Student':students_name,
    'Math':math_grades,
    'Science':science_grades,
    'English':english_grades
})
print("Student Data ---")
df.index=np.arange(1,len(df)+1) # for changing default indexing of pd which is 0
print(df.head(10))
```

### Finding top 5 Students .
#### Creating avg 

```python
df['Average']=(df['Math']+df['Science']+df['English'])/3
```

```python
top_5_students=df.sort_values(by='Average',ascending=False).head(5)
print("Top 5 Students based on Average Marks:")
print(top_5_students)
```

### Finding Subject which has Highest Class Average .

```python
subject_means = df[['Math', 'Science', 'English']].mean()
subject_with_highest_avg = subject_means.idxmax()
print("Subject with highest class average:", subject_with_highest_avg)
```

```python
plt.bar(subject_means.index, subject_means.values, color=['blue', 'green', 'red'], width=0.4 ,alpha =0.7)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.title('Average Marks by Subject')
plt.show()
```

### Finding Students with average more than 100 .

```python
students_more_than_90 = df[df['Average'] > 90]
print("Total number of Students with average more than 90 marks are ->", students_more_than_90.shape[0])
print(students_more_than_90[['Student', 'Average']])
```

### Graph representation of Average marks by each Student .

```python
plt.figure(figsize=(12, 6))  # Optional: makes the plot wider for readability
plt.bar(df['Student'], df['Average'], color='purple', alpha=0.7)
plt.xlabel('Student')
plt.ylabel('Average Marks')
plt.title('Average Marks of Each Student')
plt.xticks(rotation=45, ha='right')  # Rotates x labels for better visibility
plt.tight_layout()  # Adjusts layout to prevent label cutoff
plt.show()
```
