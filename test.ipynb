# Creating graph using numpy , matplotlib and given set of data 


### Importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

### getting random value for x_axis and y_axis 

```python
x_axis = np.arange(100)  # Sequence from 0 to 99
y_axis = np.random.randint(0, 10, 100)  # 100 random integers between 0 and 9
```

### Creating graph for the data 

```python
plt.plot(x_axis,y_axis)
plt.title("Random Scattered Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
